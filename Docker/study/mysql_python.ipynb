```python
import pymysql

# ==========================================
# [설정] 접속 정보
# ==========================================
# 1. 윈도우(Local) 실행 시: (기본값)
DB_HOST = "127.0.0.1" 

# 2. Docker(Ubuntu) 컨테이너 내부 실행 시: (주석 해제 후 사용)
# DB_HOST = "host.docker.internal" 

# Docker 실행 시 설정한 포트 (-p 3307:3306)
DB_PORT = 3307
DB_PASS = "1234"

# ------------------------------------------

# 1. 연결 설정
conn = pymysql.connect(
    host=DB_HOST, 
    user='root', 
    password=DB_PASS,
    port=DB_PORT, 
    charset='utf8mb4',
    cursorclass=pymysql.cursors.DictCursor
)

# 2. 커서 생성
with conn.cursor() as cursor:
    print(f"✅ Connected to MySQL at {DB_HOST}:{DB_PORT}")
    
    # 3. DB 생성 및 선택 (Error 1046 방지)
    cursor.execute("CREATE DATABASE IF NOT EXISTS lecture_db")
    cursor.execute("USE lecture_db")

    # 4. 테이블 생성 쿼리 실행
    # IF NOT EXISTS: 테이블이 없을 때만 생성하여 에러 방지
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS users (
            id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(50),
            score INT
        )
    """)
    
    # 5. 데이터 삽입 쿼리 실행
    cursor.execute("INSERT INTO users (name, score) VALUES ('Alice', 85)")
    cursor.execute("INSERT INTO users (name, score) VALUES ('Bob', 92)")
    
    # 커밋(Commit): 변경사항을 DB에 영구 저장 (필수!)
    conn.commit()
    
    # 6. 조회 쿼리 실행 및 결과 출력
    cursor.execute("SELECT * FROM users")
    print(cursor.fetchall()) # 모든 결과 행 가져오기

# 리소스 반환을 위해 연결 종료
conn.close()
```
